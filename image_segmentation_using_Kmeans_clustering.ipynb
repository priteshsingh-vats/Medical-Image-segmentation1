```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
im = cv2.imread("./Desktop/medical.png")
plt.imshow(im)
im = cv2.cvtColor(im,cv2.COLOR_BGR2RGB)
plt.imshow(im)
#print(im.shape)
original_img = (211, 186, 3)
plt.show()
from sklearn.cluster import KMeans
p =8
km = KMeans(n_clusters = p)
#print(im.shape)
im = im.reshape((-1,3))
#print(im.shape)
#print(im)
#print(im.shape)
km.fit(im)
centers = np.array([[123,234,14],
                    [333,331,52],
                    [255,244,23],
                    [424,324,0],[255,255,51],[224,145,0],[255,123,56],[123,214,0]])
new_img  = np.ones((211*186,3),dtype ='uint8')
#print(centers.shape)
#print(new_img)
#print(km.labels_)
#print(km.labels_.shape)
for ix in range(new_img.shape[0]):
    new_img[ix] = centers[km.labels_[ix]]
new_img = new_img.reshape(original_img)
plt.imshow(new_img)
plt.show()
```
